# JupyterLab Installation

With the prerequisites satisfied before we can start our first local PowerShell Jupyter Notebook, we need to;

* Install .NET Interactive
* Install JupyterLab
* Register .NET Interactive as a Jupyter Kernel

## Install .NET Interactive

With the .NET SDK 3.1 installed, I installed .NET Interactive from a command prompt;

```powershell
dotnet tool install -g Microsoft.dotnet-interactive
```

## Install JupyterLab

To install JupyterLab we use Python PIP. From an Administrative PowerShell command line run;

```powershell
pip install jupyterlab
```

### Register .NET Interactive as a Jupyter Kernel

Finally, we need to register .NET Interactive as a Jupyter Kernel. From an Administrative command prompt run;

```powershell
dotnet interactive jupyter install
```

# Starting Jupyter

We are now ready to start Jupyter Notebook. From an administrative PowerShell command line, run;

```powershell
jupyter notebook
```

which will start JupyterLab and open a browser window to your `localhost:8888` displaying the directory/folder we have just launched from.

## Importing PowerShell Modules

We of course want to interact with Azure, which will require some PowerShell Modules, the first and foremost is going to be the **Az** PowerShell Module. 

> NOTE: As per the announcement blog post currently there is no ability to read from the host, so the -Force switch must be used when installing modules.

> NOTE: Keep in mind .NET Interactive utilises PowerShell 6/7 so you won’t be able to use PowerShell Modules that require Windows PowerShell. Anything that previously worked with PowerShell Core 6.x should be good. 

````powershell
Install-Module -Name Az -Force
````

With the Az Module now installed it can be loaded and we can see the Az cmdlets that are available.

````powershell
Import-Module -Name Az
get-command
````

We can use the Az module Connect-AzAccount cmdlet to authenticate to Azure.

````powershell
Connect-AzAccount
````

We are now authenticated to Azure from within our PowerShell Jupyter Notebook.

We can now use other Az cmdlets to interact with Azure. Using `Get-AzVM` will list our Azure Virtual Machines. Using tab we can see `Get-AzVm` options and switches.

`Get-AzVM -Name [<vmName>]` or use tab completion to see other cmdlet options.

````powershell
Get-AzVM
````

## Key Vault Example



````powershell
$spName = "jupyter-sp"
$secretName = "jupyterClientSecret"
$kvName = "azsec-kv"
$rgName = "azsec-corporate-rg"
$kvLocation = "westus"
$userPrincipalName = "andy@azsec.net"

$sp = New-AzADServicePrincipal -Role Reader `
                               -DisplayName $spName

$spCredential = Get-AzADSpCredential -ObjectId $sp.Id

$kv = New-AzKeyVault -Name $kvName `
                     -ResourceGroupName $rgName `
                     -Location $kvLocation `

$accessPolicy = Set-AzKeyVaultAccessPolicy -VaultName $kv.VaultName `
                                           -UserPrincipalName $userPrincipalName `
                                           -PermissionsToSecrets Get,List,Set `
                                           -PassThru

if(!$accessPolicy) {
   Throw "[!] Failed to set Key Vault access policy"
}

$secretValue = ConvertTo-SecureString -String $spCredential.KeyId -AsPlainText -Force
$secret = Set-AzKeyVaultSecret -VaultName $kv.VaultName `
                               -Name $secretName `
                               -SecretValue $secretValue `

if ($secret) {
    Write-Host -ForegroundColor Green "[-] A new secret is created succesfully"
}
elseif (!$secret) {
    Write-Host -ForegroundColor Red "[!] Failed to create a secret"
}
````
